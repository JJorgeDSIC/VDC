```python
import numpy as np
import bokeh.io

bokeh.io.output_notebook()

num_series = 4
num_samples = 2000

data = np.zeros((num_series,num_samples))

for i in range(1,5):
    lines = open("./lab1-datos/v"+str(i)+".txt","r").readlines()
    
    for j,l in enumerate(lines):
        data[i-1,j] = float(l)

#print(data)

TOOLS = ""

from bokeh.layouts import gridplot,column
from bokeh.plotting import figure, show, output_file
from bokeh.io import export_png

p1 = figure(x_axis_type="linear", y_range=(4.4, 5.6))
p1.grid.grid_line_alpha=0.3
p1.xaxis.axis_label = 'Time (msecs)'
p1.yaxis.axis_label = 'Voltage (V)'

p2 = figure(x_axis_type="linear", y_range=(4.4, 5.6),tools=TOOLS)
p2.grid.grid_line_alpha=0.3
p2.xaxis.axis_label = 'Time (msecs)'
p2.yaxis.axis_label = 'Voltage (V)'

p3 = figure(x_axis_type="linear", y_range=(4.4, 5.6),tools=TOOLS)
p3.grid.grid_line_alpha=0.3
p3.xaxis.axis_label = 'Time (msecs)'
p3.yaxis.axis_label = 'Voltage (V)'

p4 = figure(x_axis_type="linear", y_range=(4.4, 5.6),tools=TOOLS)
p4.grid.grid_line_alpha=0.3
p4.xaxis.axis_label = 'Time (msecs)'
p4.yaxis.axis_label = 'Voltage (V)'

y_axis = np.arange(2000)

y_axis = y_axis * 2.5

#p1.line(y_axis, data[0], color='#A6CEE3', legend='Power Adaptor v1')
#p2.line(y_axis, data[1], color='#A6CEE3', legend='Power Adaptor v2')
#p3.line(y_axis, data[2], color='#A6CEE3', legend='Power Adaptor v3')
#p4.line(y_axis, data[3], color='#A6CEE3', legend='Power Adaptor v4')

p1.line(y_axis, data[0], color='#A6CEE3')
p2.line(y_axis, data[1], color='#A6CEE3')
p3.line(y_axis, data[2], color='#A6CEE3')
p4.line(y_axis, data[3], color='#A6CEE3')


p1.toolbar.logo = None
p1.toolbar_location = None
p2.toolbar.logo = None
p2.toolbar_location = None
p3.toolbar.logo = None
p3.toolbar_location = None
p4.toolbar.logo = None
p4.toolbar_location = None

#export_png(gridplot([[p1]], plot_width=400, plot_height=400), filename="powersV1.png")
#export_png(gridplot([[p2]], plot_width=400, plot_height=400), filename="powersV2.png")
#export_png(gridplot([[p3]], plot_width=400, plot_height=400), filename="powersV3.png")
#export_png(gridplot([[p4]], plot_width=400, plot_height=400), filename="powersV4.png")

#On file
export_png(gridplot([[p1,p2],[p3,p4]], plot_width=300, plot_height=300), filename="powers.png")
#On HTML
output_file("../../../site/PowerAdaptors.html", title="Power adaptors")
#On notebook
show(gridplot([[p1,p2],[p3,p4]], plot_width=300, plot_height=300)) 




```
